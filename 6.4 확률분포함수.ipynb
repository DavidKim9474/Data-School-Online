#### Ex. 6.4.3

```python
def P(a, b):
    if 0 <= a <= 180 and 0 <= b <= 180:
        return (abs((b - a)) * 4 / 3) / 360
    elif 0 <= a <= 180 and 180 < b <= 360:
        return (abs((180 - a)) * 4 / 3 + abs((b - 180)) * 2 / 3) / 360
    elif 180 < a <= 360 and 180 < b <= 360:
        return (abs((b - a)) * 2 / 3) / 360
    else:
        raise ValueError("Value of a and b must be between 0 and 360 (inclusive) and b must be greater than or equal to a")
```

#### 누적분포함수

```python
t = np.linspace(-100, 500, 100)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
plt.plot(t, F)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.title("누적분포함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$F(x)$")
plt.show()
```

#### Ex. 6.4.4

```python
def F(a):
    if a < 0:
        return 0
    if a > 360:
        return 1
    elif a < 180:
        return (2 / 3) * (a / 180)
    elif a >= 180:
        return (2 / 3) + (1 / 3) * ((a - 180) / 180)

t = np.linspace(-100, 500, 100)
Fs = np.array(list(map(F, t)))    # 벡터와 연산이 불가능한 것을 list(map(function, variable)로 가능하게!
plt.plot(t, Fs)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.title("누적분포함수 (왜곡된 경우)")
plt.xlabel("$x$ (도)")
plt.ylabel("$F(x)$")
plt.show()
```

#### 확률밀도함수

```python
t = np.linspace(-100, 500, 1000)
F = t / 360
F[t < 0] = 0
F[t > 360] = 1
p = np.gradient(F, 600/1000)
plt.plot(t, p)
plt.ylim(-0.0001, p.max()*1.1)
plt.xticks([0, 180, 360])
plt.title("확률밀도함수")
plt.xlabel("$x$ (도)")
plt.ylabel("$p(x)$")
plt.show()
```
