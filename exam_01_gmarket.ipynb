### Gmarket 베스트 상품 200개 크롤링
- 상품명, 원가, 판매가
- 썸네일 이미지

```python
response = requests.get("http://corners.gmarket.co.kr/Bestsellers")
dom = BeautifulSoup(response.content, "html.parser")
selector = "#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li"
elements = dom.select(selector)
len(elements)
```

```python
# 코드, 제목, 링크, 원가, 판매가, 할인율
code_selector = "#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li > div.thumb > a"
title_selector = "#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li > a.itemname"
price_selector = "#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li > div.item_price > div.s-price"
elements[2].select_one(code_selector).get("href")[41:51], elements[2].select_one(title_selector).text, elements[2].select_one(price_selector).text.strip().split("\n")[0]
```

```python
data = []
for element in elements:
    data.append({
        "code": str(element.select_one(code_selector).get("href")[41:51]).replace("&", ""),
        "title": element.select_one(title_selector).text,
        "price": element.select_one(price_selector).text.strip().split("\n")[0],
    })
items_df = pd.DataFrame(data)
items_df
```

```python
def download(title, link):
    
    response = requests.get(link, stream=True, allow_redirects=False)
    path = "data/{}".format(title)
    
    with open(path, "wb") as f:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
```

```python
for idx, row in items_df.iterrows():
    link = "http://gdimg.gmarket.co.kr/{}/still/300".format(row['code'])
    title = row['title'] + '.png'
    title = title.replace("/", "")
    print(title, link)
    download(title, link)
```
