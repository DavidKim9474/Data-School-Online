```python
# 함수로 만들기
import requests
from bs4 import BeautifulSoup

def jobKorea_crawling():
    jobKorea_url = "http://www.jobkorea.co.kr/Top100/?\
    Main_Career_Type=1&Search_Type=1&BizJobtype_Bctgr_Code=10016"
    jobKorea_response = requests.get(jobKorea_url)
    jobKorea_dom = BeautifulSoup(jobKorea_response.content, "html.parser")
    jobKorea_elements = dom.select("#devTypeTab_1 > div.rankListWrap > div.rankListArea.devSarterTab > ol > li")
    for element in jobKorea_elements:
        jobs_data.append({
            "company": element.select_one("div.co > div.coTit > a").text.strip(),
            "job_title": element.select_one("div.info > div.tit > a").text.strip(),
            "due_date" : element.select_one("div.side > span.day").text.strip(),
            "link" : 'http://www.jobkorea.co.kr' + element.select_one("div.info > div.tit > a").get("href"),
        })
    jobKorea_df = pd.DataFrame(jobKorea_data)
    return jobKorea_df
```
