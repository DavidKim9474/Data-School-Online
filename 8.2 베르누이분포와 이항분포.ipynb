#### 사이파이를 사용한 이항분포의 시뮬레이션

```python
N = 10
mu = 0.6
rv = sp.stats.binom(N, mu)
```

```python
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
N = 10
mu = 0.9
rv = sp.stats.binom(N, mu)

xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
N = 10
mu = 0.2
rv = sp.stats.binom(N, mu)

xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
N = 10
mu = 1
rv = sp.stats.binom(N, mu)

xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align="center")
plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("이항분포의 확률질량함수")
plt.show()
```

```python
N = 10
mu = 0.6
rv = sp.stats.binom(N, mu)
np.random.seed(0)
x = rv.rvs(100)
x
```

```python
sns.countplot(x)
plt.title("이항분포의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.show()
```

```python
y = np.bincount(x, minlength=N+1) / float(len(x))
df = pd.DataFrame({"이론": rv.pmf(xx), "시뮬레이션": y}).stack()
df = df.reset_index()
df.columns = ["표본값", "유형", "비율"]
df.pivot("표본값", "유형", "비율")
df
```

```python
sns.barplot(x="표본값", y="비율", hue="유형", data=df)
plt.title("이항분포의 이론적 분포와 시뮬레이션 분포")
plt.show()
```
