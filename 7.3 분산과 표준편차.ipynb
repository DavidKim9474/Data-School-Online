#### Ex. 7.3.3

```python
x1 = np.random.normal(size=100)
x1
```

```python
x1_var = x1.var()
x1_var
```

```python
x_list = []
x_list_mean = []
for i in range(50):
    x = np.random.normal(size=100)
    x_mean = x.mean()
    x_list.append(x)
    x_list_mean.append(x_mean)
```

```python
x_list
```

```python
x_list_mean
```

```python
x_mean_var = np.asarray(x_list_mean).var()
x_mean_var
```

```python
round(x1_var / x_mean_var, 2)
```

#### 표본분산의 기댓값

```python
np.random.seed(15)
N = 7
data = np.sort(np.random.normal(size=(N)))[::-1]
```

```python
mean = np.mean(data)
mean
```

```python
plt.figure(figsize=(10, 2))
sns.rugplot(data, height=0.5, linewidth=4)
x = np.linspace(-3, 3, 100)
plt.axvline(x=0, ls=":", c="r", linewidth=2, label="실제 기댓값")
plt.axvline(x=mean, ls="--", c="b", linewidth=2, label="표본평균")
plt.legend()
plt.xlim(-2, 2)
plt.show()
```

```python
distance_from_sample_mean = data - mean
distance_from_sample_mean
```

```python
sample_variance = (distance_from_sample_mean ** 2).mean()
sample_variance
```

```python
sample_variance * N / (N - 1)
```
